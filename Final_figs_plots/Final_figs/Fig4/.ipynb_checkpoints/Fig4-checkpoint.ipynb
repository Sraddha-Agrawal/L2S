```python
from ase.io import read
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d, InterpolatedUnivariateSpline
from scipy.integrate import solve_ivp
from scipy.optimize import minimize
from pathlib import Path
```

Fig 1a

```python
energies = np.loadtxt("Data/toten.dat")
energies -= np.min(energies)
x_saddle=0.23589477841686
x = np.linspace(-0.5, 0.9, len(energies))
x=x-x_saddle # to position the saddle point at x=zero
xfine = np.linspace(-0.5, 0.9, 1000)
xfine=xfine-x_saddle # to position the saddle point at x=zero
x_coarse = np.loadtxt("Data/x_1_a2.dat")
Rx_coarse = np.loadtxt("Data/Rx_coarse_1_a2.dat")
Rx_fine = np.loadtxt("Data/Rx_fine_1_a2.dat")

fig1a, (ax0,ax1) = plt.subplots(2, 1, figsize=(2, 3), dpi=150, facecolor='white', sharex=True)

ax0.plot(x, energies, 'C0o-', markersize=1.5)
ax0.set_xlim(-0.45, 0.45)
ax0.set_ylim(-0.05, 0.275)
ax0.set_ylabel('Energy (eV)')
#ax1.set_xlabel('x (Å)')

ax1.scatter(x_coarse, Rx_coarse, s=8, color='C0')
ax1.plot(xfine, Rx_fine)
ax1.set_xlim(-0.45, 0.45)
ax1.set_ylim(-150, 150)
ax1.set_xlabel("x (Å)")
ax1.set_ylabel("R(x)")

ax0.tick_params(direction='in', labelbottom=False)
ax1.tick_params(direction='in')
ax0.text(-0.03, 1.2, "(a)", transform=ax0.transAxes,fontsize=14, va='top', ha='right')

plt.subplots_adjust(hspace=0)
#fig.tight_layout()
```

```python
# Fig 1b
from matplotlib.colors import LinearSegmentedColormap, Normalize
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

freq_in_thz = 6.3507799295888985 # converts w=sqrt(k/m) in sqrt(THz/(Angstrom^2 * atomic mass unit)) to THz
t_in_ps = 1 / freq_in_thz

tspan = [0, 20/t_in_ps]
tfine = np.linspace(*tspan, 20000)

sol_x = np.loadtxt('Data/sol_x_1b.dat')
fft_x = np.loadtxt('Data/fft_x_1b.dat')
freq_x = np.loadtxt('Data/freq_x_1b.dat')

fig1b, ax2 = plt.subplots(figsize=(3.5, 3), dpi=150, facecolor='white')
ax2_inset = inset_axes(ax2, width="40%", height="35%", loc='upper right', borderpad=1)
cmap = plt.get_cmap('Dark2')
colors = [cmap(i) for i in range(3)]

# Create two custom colormaps for the two segments
segment1_cmap = LinearSegmentedColormap.from_list('segment1', [colors[0], colors[1]], N=15)
segment2_cmap = LinearSegmentedColormap.from_list('segment2', [colors[1], colors[2]], N=26)
combined_colors = np.vstack((segment1_cmap(np.linspace(0, 1, 15)),
                             segment2_cmap(np.linspace(0, 1, 26))))

for i in range(sol_x.shape[0]):
    ax2.plot(tfine*t_in_ps, sol_x[i], color=colors[i], alpha = 0.8)
ax2.set_xlabel("Time (ps)", labelpad=5)
ax2.set_ylabel("x (Å)")
ax2.set_xlim(0,5)

for i in range(fft_x.shape[0]):
    ax2_inset.plot(freq_x, fft_x[i]+(i*0.035), color=combined_colors[i], alpha=0.48)
    
ax2_inset.set_xlabel('ω (THz)')
#ax2_inset.set_ylabel('Amplitude')
ax2.tick_params(direction='in')
ax2_inset.tick_params(direction='in')
ax2_inset.set_xlim(0.01,6.5)
ax2.text(-0.03, 1.1, "(b)", transform=ax2.transAxes,fontsize=14, va='top', ha='right')
```

```python
# Fig 1c
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import FormatStrFormatter
    
sol_x_sine = np.loadtxt('Data/sol_x_1c.dat')
sol_v_sine = np.loadtxt('Data/sol_v_1c.dat')

protocol = np.loadtxt('Data/protocol_1c.dat')
norm_protocol = np.loadtxt('Data/norm_protocol_1c.dat')

fig1c, ax3 = plt.subplots(figsize=(3.5, 3), dpi=150, facecolor='white')
ax3_inset = inset_axes(ax3, width="40%", height="35%", loc='upper right', borderpad=1)
norm1 = Normalize(vmin=np.min(tfine), vmax=np.max(tfine*t_in_ps))

ax3.scatter(sol_x_sine, sol_v_sine, c = tfine*t_in_ps, cmap='PiYG', edgecolor='none', norm=norm1, s=8)
ax3_inset.plot(tfine, protocol)
ax3.set_xlabel('x')
ax3.set_ylabel('v')
ax3.tick_params(direction='in')
ax3_inset.set_xlabel('Time(ps)')
ax3_inset.set_ylabel('sin(ωt)')
ax3.set_xlim(-0.237,-0.2358)
ax3.set_ylim(-0.0015,0.0025)
formatter = ScalarFormatter(useOffset=True, useMathText=True)
formatter.set_powerlimits((-2, 2))
ax3.yaxis.set_major_formatter(formatter)
#ax3.xaxis.set_major_formatter(formatter)
ax3.xaxis.set_major_formatter(FormatStrFormatter('%.3f'))
ax3_inset.set_xlim(0,10)
ax3_inset.tick_params(direction='in')
ax3.text(-0.03, 1.1, "(c)", transform=ax3.transAxes,fontsize=14, va='top', ha='right')
```
